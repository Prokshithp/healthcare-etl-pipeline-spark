```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Healthcare Data Lake ETL") \
    .getOrCreate()
```

```python
df = spark.read.csv(
    "/content/patients_raw.csv",
    header=True,
    inferSchema=True
)

df.printSchema()
df.show(5)
print("Total records:", df.count())
```
